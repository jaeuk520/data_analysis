```python
import csv
import matplotlib.pyplot as plt
import matplotlib

def main():
    f = open('2.약국정보서비스의 복사본.csv','r')
    data = csv.reader(f)
    next(data)
    cities = []
    cities_cnt = dict()
    for row in data:
        city = row[5]
        if city not in cities:
            cities.append(city)
            cities_cnt[city] = 1
        else:
            cnt = cities_cnt.get(city) + 1
            cities_cnt[city] = cnt
    print(cities)
    print(cities_cnt)

    ratio = list(cities_cnt.values())
    explode = [0.10] * 17

    #한글 깨짐 해결
    matplotlib.rcParams['font.family'] ='Malgun Gothic'
    matplotlib.rcParams['axes.unicode_minus'] =False
    
    plt.pie(ratio, labels=cities, autopct='%.1f%%', explode = explode)

if __name__=="__main__":
    main()

```
