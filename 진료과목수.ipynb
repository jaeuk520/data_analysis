```python
import csv
import matplotlib.pyplot as plt
import matplotlib
import operator
import numpy as np

def main():
    f = open('5.의료기관별상세정보서비스_03_진료과목정보의 복사본.csv','r')
    data = csv.reader(f)
    next(data)
    subjects = []
    subjects_cnt = dict()
    for row in data:
        subject = row[3]
        if subject not in subjects:
            subjects.append(subject)
            subjects_cnt[subject] = 1
        else:
            cnt = subjects_cnt.get(subject) + 1
            subjects_cnt[subject] = cnt

    s_sorted = dict(sorted(subjects_cnt.items(), key=operator.itemgetter(1))[0:9])
    s_sorted_reverse = dict(sorted(subjects_cnt.items(), key=operator.itemgetter(1), reverse=True)[0:9])
    
    print(s_sorted)
    print(s_sorted_reverse)
    
    x = np.arange(9)
    
    plt.subplot(2,1,1)
    plt.bar(x, s_sorted.values(), width=0.6)
    plt.title('진료과목수 하위 9')
    plt.xticks(x, s_sorted.keys(), rotation=90)

    plt.subplot(2,1,2)
    plt.bar(x, s_sorted_reverse.values(), width=0.6)
    plt.title('진료과목수 상위 9')
    plt.xticks(x, s_sorted_reverse.keys(), rotation=90)

    plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=1.5, wspace=0.2, hspace=0.8)
    
    plt.show()

if __name__=="__main__":
    main()

```
