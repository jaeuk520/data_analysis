```python
import csv
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import operator

def main():
    f = open('1.병원정보서비스의 복사본.csv','r')
    data = csv.reader(f)
    next(data)
    hospitals = dict()
    for row in data:
        cnt_doctor = int(row[14])
        hospital = row[1]
        hospitals[hospital] = cnt_doctor

    #print(hospitals)
    h_sorted = sorted(hospitals.items(), key=operator.itemgetter(1), reverse=True)[0:10]
    print(h_sorted)
    
    #한글 깨짐 해결
    matplotlib.rcParams['font.family'] ='Malgun Gothic'
    matplotlib.rcParams['axes.unicode_minus'] =False

    x = np.arange(10)
    values=[]
    names=[]
    for i in range(10):
        values.append(h_sorted[i][1])
        names.append(h_sorted[i][0])
    plt.bar(x, values, width=0.6)
    plt.xticks(x, names, rotation=90)

    plt.show()

if __name__=="__main__":
    main()

```
